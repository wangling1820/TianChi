```python
def linear_regression_equation(data):
    y_mean = data['y'].mean()
    x_mean = data['x'].mean()
    x_2 = data['x'].map(lambda x:  x * x - x_mean * x_mean)
    x_y = data.apply(lambda x: x['x'] * x['y'] - x_mean * y_mean, axis=1)
    b = x_y.sum() /  x_2.sum()
    a = y_mean - x_mean * b
    return a, b
```
