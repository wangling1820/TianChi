```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras import optimizers
from keras.layers import Dropout
```

```python
import tensorflow as tf
tf.__version__
import keras as k
k
```

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('./data/train_feature.csv')
train.head()
```

```python
import numpy as np
label_dict = {'拖网':0, '刺网':1, '围网':2}
train['type'] = train['type'].map(label_dict)
label = np.array(train['type'])
label
```

```python
one_hot_label = to_categorical(label)
one_hot_label
```

```python
x = np.array(train.drop('type', axis=1))
x.shape
```

```python
x
```

```python
def get_model():
    model = Sequential()
    model.add(Dense(128, input_dim=39, activation='relu'))
    model.add(Dense(64,  activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(3, activation='softmax'))
    sgd = optimizers.SGD(lr=0.01, decay=1e-4, momentum=0.9, nesterov=True)
    model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
    return model
```

```python
my_model = get_model()
```

```python
my_model.fit(x=x, y=one_hot_label, batch_size=256, epochs=200, validation_split=0.15, shuffle=True)
```

```python
test_fea = pd.read_csv('./data/test_feature.csv')
test_fea
test = np.array(test_fea.drop('id', axis=1))
```

```python
result = my_model.predict(test)
```
